```python
import pandas as pd

a = pd.read_csv('aitimes_articles_context.csv')
a
```

```python
from TEAM.HybridModel import StrategyCSummarizer

# 1. 10개 데이터만 슬라이싱하여 테스트용 데이터프레임 복사
num = 50
test_df = a.sample(n=num).copy()

# 2. 하이브리드 요약기(전략 C) 인스턴스 생성
# HybridModel.py에서 사용한 파라미터 값 적용 (beta_param=0.5: BM25와 LSA 1:1 반영)
summarizer = StrategyCSummarizer(lambda_param=0.7, alpha_param=0.2, beta_param=0.5, n_components=3)

# 3. 요약 결과를 담을 리스트 설정 (5문장 요약)
TARGET_SENTENCES = 3
summaries = []

print(f"데이터 {num}개에 대한 {TARGET_SENTENCES}문장 요악을 진행 중입니다...")

for idx, row in test_df.iterrows():
    text = str(row['context'])
    
    # 텍스트가 존재하는 경우에만 요약 수행
    if text.strip():
        try:
            summary = summarizer.summarize(text, top_k=TARGET_SENTENCES)
            summaries.append(summary)
        except Exception as e:
            print(f"Index {idx} 요약 중 에러 발생: {e}")
            summaries.append("") # 에러 시 빈 문자열 처리
    else:
        summaries.append("")

# 4. 데이터프레임에 요약 결과 컬럼 추가
test_df['hybrid_summary'] = summaries

print("요약 완료!\n")

# 5. 데이터프레임에서 '제목', '원문', '요약문' 비교 출력 (디스플레이용)
b = test_df[['title', 'context', 'hybrid_summary']]
b.head()
```

```python
import pandas as pd
from tqdm.auto import tqdm
import ollama

# 요약 함수 정의
def summarize_with_gemma(text):
    # 예외 처리: 텍스트가 비어있거나 문자열이 아닌 경우
    if not isinstance(text, str) or not text.strip():
        return ""
        
    prompt = f"""다음 기사를 정확히 3개의 문장으로 요약하라.
[출력 규칙]
- 반드시 3줄만 출력한다.
- 각 줄은 하나의 완전한 문장만 포함한다.
- 줄 구분은 단일 개행 문자(\n)만 사용한다.
- 줄 사이에 빈 줄을 절대 넣지 않는다.
- 번호, 기호, 불릿, 설명을 절대 포함하지 않는다.
- 첫 줄 앞과 마지막 줄 뒤에 공백이나 개행을 넣지 않는다.
- 출력은 오직 요약문 3줄만 작성한다.
{text}"""

    try:
        # ollama gemma3:27b 모델 호출
        response = ollama.chat(
            model='gemma3:27b', 
            messages=[{'role': 'user', 'content': prompt}]
        )
        return response['message']['content'].strip()
    except Exception as e:
        print(f"요약 중 오류 발생 (내용 일부: {text[:30]}...): {e}")
        return None  # 에러 발생 시 처리

# pandas 데이터프레임에 tqdm 적용
tqdm.pandas(desc="Gemma3 요약 진행 중")

# b['context']의 모든 열에 요약 함수 적용 후 새로운 컬럼 할당
b['gemma3_summary'] = b['context'].progress_apply(summarize_with_gemma)

# 결과 확인
display(b[['title', 'context', 'hybrid_summary', 'gemma3_summary']].head())
```

```python
import torch
import numpy as np
from rouge_score import rouge_scorer
from bert_score import BERTScorer

# 1. 디바이스 설정 및 모델 로딩
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device} for BERTScore")

# ROUGE 및 BERT 스코어러 초기화 (analysis.py와 동일 설정)
rouge_evaluator = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
bert_scorer = BERTScorer(lang="ko", rescale_with_baseline=False, device=device)

# 2. 유효한 데이터 필터링 
# (정답이나 평가 대상이 비어있으면 스코어링에서 제외하기 위해 복사본 사용)
subset = b.dropna(subset=['gemma3_summary', 'hybrid_summary']).copy()
subset = subset[(subset['gemma3_summary'] != "") & (subset['hybrid_summary'] != "")]

refs = subset['gemma3_summary'].tolist()   # 정답 (Ground Truth)
preds = subset['hybrid_summary'].tolist()  # 평가 대상 (Prediction)

print(f"\n평가 대상 데이터 개수: {len(refs)}개\n")

if len(refs) > 0:
    # 3. ROUGE 점수 계산 (row 단위)
    r1_f1, r2_f1, rl_f1 = [], [], []
    for r, p in zip(refs, preds):
        score = rouge_evaluator.score(r, p)
        r1_f1.append(score['rouge1'].fmeasure)
        r2_f1.append(score['rouge2'].fmeasure)
        r1_f1.append(score['rougeL'].fmeasure)
        
        # 참고: rouge1, rougeL은 위와 같이 append하지만, 통일성을 위해 List에 모두 추가
        # (위에 오타 방지를 위해 명확히 변수 분리)
        # correction:
    
    r1_f1, r2_f1, rl_f1 = [], [], []
    for r, p in zip(refs, preds):
        score = rouge_evaluator.score(r, p)
        r1_f1.append(score['rouge1'].fmeasure)
        r2_f1.append(score['rouge2'].fmeasure)
        rl_f1.append(score['rougeL'].fmeasure)
        
    subset['rouge1_f1'] = r1_f1
    subset['rouge2_f1'] = r2_f1
    subset['rougeL_f1'] = rl_f1

    # 4. BERT 점수 계산 (배치로 한꺼번에 처리)
    # bert_scorer.score()는 각 문장 쌍에 대해 P(정밀도), R(재현율), F1(F1 점수) 텐서를 반환함
    P, R, F1 = bert_scorer.score(preds, refs)
    subset['bert_p'] = P.cpu().numpy()
    subset['bert_r'] = R.cpu().numpy()
    subset['bert_f1'] = F1.cpu().numpy()

    # 5. 원본 데이터프레임 `b`에 점수 병합
    score_cols = ['rouge1_f1', 'rouge2_f1', 'rougeL_f1', 'bert_p', 'bert_r', 'bert_f1']
    for col in score_cols:
        b[col] = subset[col]

    # 6. 결과 리포트 출력
    print("="*60)
    print("평가 결과 요약 (Ground Truth: Gemma3 / Pred: Hybrid)")
    print("="*60)
    print(f"ROUGE-1 F1 Mean : {subset['rouge1_f1'].mean():.4f}")
    print(f"ROUGE-2 F1 Mean : {subset['rouge2_f1'].mean():.4f}")
    print(f"ROUGE-L F1 Mean : {subset['rougeL_f1'].mean():.4f}")
    print(f"BERT-P Mean     : {subset['bert_p'].mean():.4f}")
    print(f"BERT-R Mean     : {subset['bert_r'].mean():.4f}")
    print(f"BERT-F1 Mean    : {subset['bert_f1'].mean():.4f}")
    print("="*60)
else:
    print("평가할 대상 문장쌍(결측치 제외)이 존재하지 않습니다.")

# 개별 행별 점수 확인 (앞쪽 5개)
display(b[['gemma3_summary', 'hybrid_summary', 'rouge1_f1', 'bert_f1']].head())
```
